```python
from models.XGB.xgb_model import XGBModel

import os
import pandas as pd

ROOT_PATH = os.path.pardir
PLOTS_DIR = os.path.join(ROOT_PATH, "plots")

SUNSPOTS_PATH = os.path.join(ROOT_PATH, "datasets/Sunspots.csv")
MACKEY_GLASS_PATH = os.path.join(ROOT_PATH, "datasets/MackeyGlass.csv")
ELECTRICITY_PATH = os.path.join(ROOT_PATH, "datasets/Electricity.csv")
```

```python
XGBModelSunspots = XGBModel(
    dataset=pd.read_csv(SUNSPOTS_PATH),
    n_steps_in=10,
    n_steps_out=10,
    test_frac=0.1,
    metric=None,
    xlabel="Month Index",
    ylabel="Monthly Mean Total Sunspot Number",
    target_column="Monthly Mean Total Sunspot Number"
)
```

```python
XGBModelElectricity = XGBModel(
    dataset=pd.read_csv(ELECTRICITY_PATH),
    n_steps_in=20,
    n_steps_out=10,
    test_frac=0.1,
    metric=None,
    xlabel="Day Index",
    ylabel="TODO",
    target_column="IPG2211A2N"
)
```
